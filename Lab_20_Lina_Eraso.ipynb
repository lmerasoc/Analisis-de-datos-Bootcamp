```python
df = pd.read_csv('datos_financieros.csv')
print(df.count())
print(df.head(20))
```

```python
print(df.info())
print(df.describe(include='all'))
```

```python
#Valores nulos
print(df.isnull().sum())
```

```python
#Duplicated
print("Datos duplicados:",df.duplicated().sum())
print(df.duplicated())
```

```python
#Llenar datos nulos
df1=df
df1['Salario_Anual'] = df1['Salario_Anual'].fillna(df1['Salario_Anual'].mean())
df1['Edad'] = df1['Edad'].fillna(df1['Edad'].mean())
print(df1.isnull().sum())
print(df1.count())
```

```python
#Eliminar datos
df2 = df1.dropna()
print(df2.isnull().sum())
print(df2.count())
```

```python
#Eliminar duplicados
df3 = df2.drop_duplicates()
df4=df3
print("Datos duplicados:",df4.duplicated().sum())
print(df3.count())
```

```python
print(df4.info())
print(df4.describe(include='all'))
```

```python
#exportar
df4.to_csv('datos_financieros_limpio', index=False)
print(df4)
from google.colab import files
files.download('datos_financieros_limpio')
```
