Parte 1: Introducción a Python
-	Ejercicio breve: Escribir un programa que reciba una lista de números enteros y determine la cantidad de números pares e impares.


```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd

```

```python
Lista = np.random.randint(0,100, size=10).tolist()
contadorp=0
contadori=0
for numero in Lista:
  if numero % 2 == 0:
    contadorp=contadorp+1
  else:
    contadori=contadori+1
print("Lista:",Lista)
print("Numeros pares:",contadorp)
print("Numeros impares:",contadori)
```

Parte 3: Visualización de datos con Matplotlib y Seaborn
Gráficos básicos con Matplotlib:

•	Crear un gráfico de barras que agrupe las ventas por sucursal y mes, y sume las cantidades vendidas.

•	Crear un gráfico de dispersión para mostrar la relación entre las columnas “Precio Unitario” y “Cantidad Vendida”.


```python
df_sum = df1.groupby(['Sucursal', 'Mes'])['Cantidad_vendida'].sum().reset_index()
sns.set_style("whitegrid")
plt.figure(figsize=(12, 6))
sns.barplot(data=df_sum, x='Sucursal', y='Cantidad_vendida', hue='Mes', palette='pastel', width=0.7)
plt.title('Ventas por Sucursal y Mes')
plt.xlabel('Sucursal')
plt.ylabel('Cantidad Vendida')
plt.legend(title='Mes')
plt.tight_layout()
plt.show()
```

```python
df_sum = df1.groupby(['Sucursal', 'Mes'])['Cantidad_vendida'].sum().reset_index()
df_pivot = df_sum.pivot(index='Sucursal', columns='Mes', values='Cantidad_vendida').fillna(0)
n = len(df_pivot.index)
x = np.arange(n)
width = 0.2
plt.figure(figsize=(12, 6))
meses = ['Enero', 'Febrero', 'Marzo','Abril']
plt.bar(x - 1.5* width, df_pivot['Enero'], width=width, label=meses[0])
plt.bar(x - 0.5*width, df_pivot['Febrero'], width=width, label=meses[1])
plt.bar(x + 0.5*width, df_pivot['Marzo'], width=width, label=meses[2])
plt.bar(x + 1.5*width, df_pivot['Abril'], width=width, label=meses[3])
plt.xticks(x, df_pivot.index)
plt.title('Ventas por Sucursal y Mes')
plt.xlabel('Sucursal')
plt.ylabel('Cantidad Vendida')
plt.legend(title='Mes')
plt.tight_layout()
plt.show()
```
