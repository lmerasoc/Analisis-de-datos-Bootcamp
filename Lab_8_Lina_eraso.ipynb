-	Realizar un gráfico personalizado (aplicar un estilo atractivo) donde se pueda analizar las ventas de los 8 productos (Producto A, Producto B y Producto C, ingresar los que faltan) en tres ciudades (Bogotá, Cali, Bucaramanga).

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
df1=pd.DataFrame({
    'Producto': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']*3,
    'ciudades': ['Bogotá', 'Cali', 'Bucaramanga']*8,
   'ventas':[120, 95, 130, 85, 100, 70, 110, 90,# Bogotá
    100, 80, 105, 75, 95, 65, 120, 85,# Cali
    90, 70, 100, 60, 85, 55, 95, 75]# Bucaramanga

})
df1
```

```python
df = pd.DataFrame({
    'Producto': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H'],
    'Bogotá': [120, 95, 130, 85, 100, 70, 110, 90],
    'Cali': [100, 80, 105, 75, 95, 65, 120, 85],
    'Bucaramanga': [90, 70, 100, 60, 85, 55, 95, 75]
})
df
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(12, 6))
sns.barplot(data=df1, x='Producto', y='ventas', hue='ciudades', palette='pastel')
plt.title('Ventas por Producto en Tres Ciudades')
plt.xlabel('Producto')
plt.ylabel('Ventas')
plt.legend(title='Ciudad')
plt.tight_layout()
plt.show()
```

-	Realizar un gráfico personalizado (aplicar un estilo atractivo) para visualizar la distribución de las puntuaciones en un examen según el nivel de motivación de los estudiantes, utilizando histogramas separados para baja, media y alta motivación.

```python
df4=pd.DataFrame({
    'Puntuación en el examen': [50,60,65,70,75,78,80,82,85,87,90,92,94,96,98,99,
                                100,102,103,105],
    'Nivel de motivación': ['Baja', 'Media', 'Media', 'Alta', 'Alta', 'Media',
                            'Media', 'Alta', 'Alta', 'Baja', 'Alta', 'Alta', 'Media', 'Alta', 'Alta',
                            'Media', 'Alta', 'Media', 'Baja','Alta']
})
df4
```

```python
Alta=df4[df4['Nivel de motivación']=='Alta']['Puntuación en el examen']
bins_sturges1 = int(np.ceil(np.log2(len(Alta)) + 1))
Media=df4[df4['Nivel de motivación']=='Media']['Puntuación en el examen']
bins_sturges2 = int(np.ceil(np.log2(len(Media)) + 1))
Baja=df4[df4['Nivel de motivación']=='Baja']['Puntuación en el examen']
bins_sturges3 = int(np.ceil(np.log2(len(Baja)) + 1))
fig, (ax0, ax1, ax2) = plt.subplots(1, 3, figsize=(15, 6))
colors = ['#fbb4b9', '#b3cde0', '#decbe4']
ax0.hist(Alta, bins=bins_sturges1, density=True, histtype='bar', color=colors[0], label='Alta')
ax1.hist(Media, bins=bins_sturges2, density=True, histtype='bar', color=colors[1], label='Media')
ax2.hist(Baja, bins=bins_sturges3, density=True, histtype='bar', color=colors[2], label='Baja')
ax0.legend(prop={'size': 10})
ax0.set_title('Motivación Alta')
ax1.legend(prop={'size': 10})
ax1.set_title('Motivación Media')
ax2.legend(prop={'size': 10})
ax2.set_title('Motivación baja')

```

-	Realizar un gráfico personalizado para aplicar un estilo atractivo, se requiere visualizar las ventas mensuales de tres tipos de vehículos (SUV, Sedan, y Hatchback) durante un año.

```python
df5=pd.DataFrame({
    'Mes': ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio',
            'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre']*3,
    'Vehículo': ['SUV']*12+['Sedan']*12+['Hatchback']*12,
   'ventas':[120,130,135,140,150,160,170,180,175,185,190,200, #SUV
   100, 90,95,110,120,130,140,145,150,160,170,180, # Sedan
   80,85,90,95,100,105,110,115,120,125,130,135]# Hatchback

})
df5
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(12, 6))
sns.lineplot(data=df5, x='Mes', y='ventas', hue='Vehículo', palette='Set1')
plt.title('Ventas por mes de tres vehiculos')
plt.xlabel('Mes')
plt.ylabel('Ventas')
plt.xticks(rotation=45)
plt.legend(title='Vehículo')
plt.tight_layout()
plt.show()
```

```python
df6=pd.DataFrame({
    'Mes': ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio',
            'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre'],
    'SUV':[120,130,135,140,150,160,170,180,175,185,190,200],
    'Sedan':[100, 90,95,110,120,130,140,145,150,160,170,180],
    'Hatchback:':[80,85,90,95,100,105,110,115,120,125,130,135]
})
df6
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Crear el DataFrame con los datos de ventas mensuales por tipo de vehículo

datos = pd.DataFrame({
'Mes': ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre'],
'SUV': [120, 130, 135, 140, 150, 160, 170, 180, 175, 185, 190, 200],
'Sedan': [100, 90, 95, 110, 120, 130, 140, 145, 150, 160, 170, 180],
'Hatchback': [80, 85, 90, 95, 100, 105, 110, 115, 120, 125, 130, 135]
})

# Configurar el estilo de Seaborn
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))

# Crear el gráfico de líneas con Matplotlib
plt.plot(datos['Mes'], datos['SUV'], label='SUV', marker='o', color='blue', linestyle='--')
plt.plot(datos['Mes'], datos['Sedan'], label='Sedan', marker='s', color='green', linestyle='-.')
plt.plot(datos['Mes'],	datos['Hatchback'],	label='Hatchback',	marker='^',	color='red', linestyle=':')

# Personalizar el gráfico con títulos y etiquetas
plt.title('Ventas Mensuales de Vehículos por Tipo', fontsize=14)
plt.xlabel('Meses', fontsize=12)
plt.ylabel('Ventas (unidades)', fontsize=12)

plt.legend()
plt.tight_layout()
plt.show()
```
